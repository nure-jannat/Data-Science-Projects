# Australian Credit Approval using Support Vector Machines (SVMs) 
 
### [Statlog (Australian Credit Approval) Data Set ](https://archive.ics.uci.edu/ml/datasets/Statlog+%28Australian+Credit+Approval%29)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Aust_Credit_Approval_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Let's move on to machine learning. 
## Machine Learning 

```python
X = df.iloc[:,:-1]
y = df['target']
```

```python
print(y)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

### Importing and  training the Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear', C=30, gamma='auto')
```

```python
svm_model.fit(X_train,y_train)
```

```python
predictions = svm_model.predict(X_test)
predictions
```

```python
svm_model.score(X_test, y_test)
```
